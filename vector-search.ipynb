```python
raw_documents = TextLoader(
    "tagged_description.txt",encoding="utf-8"
).load()

text_splitter = CharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    separator="\n"
)

documents = text_splitter.split_documents(raw_documents)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from sentence_transformers import SentenceTransformer

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)


db_books = Chroma.from_documents(
    documents,
    embedding=embeddings
)
```

```python
query = "A book to teach children about nature"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
books[books["isbn13"]==int(docs[0].page_content.split()[0].strip('":'))]
```

```python
def retrieve_semantic_recommendations(
        query:str,
        top_k:int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query,k=50)
    books_list = []
    for i in range(0,len(recs)):
        books_list += [int(recs[i].page_content.split()[0].strip('":'))]
    return books[books["isbn13"].isin(books_list)].head(top_k)

```

```python
import numpy as np
max_index =np.argmax(pipe(sequence,fiction_categories)["scores"])
max_label= pipe(sequence,fiction_categories)["labels"][max_index]
max_label
```

```python
def generate_predictions(sequence,categories):
    predictions = pipe(sequence,categories)
    max_index = np.argmax(predictions["scores"])
    max_label = predictions["labels"][max_index]
    return max_label
```

```python
from tqdm import tqdm
actual_categories = []
predicted_categories =[]

fiction_desc = books.loc[
    books["simple_categories"] == "Fiction",
    "description"
].reset_index(drop=True)

for i in tqdm(range(300)):
    sequence = fiction_desc[i]
    predicted_categories.append(
        generate_predictions(sequence, fiction_categories)
    )
    actual_categories.append("Fiction")
```

```python
nonfiction_desc = books.loc[
    books["simple_categories"] == "Nonfiction",
    "description"
].reset_index(drop=True)

for i in tqdm(range(300)):
    sequence = nonfiction_desc[i]
    predicted_categories.append(
        generate_predictions(sequence, fiction_categories)
    )
    actual_categories.append("Nonfiction")
```

```python
isbns =[]
predicted_categories=[]
missing_categories =books.loc[books["simple_categories"].isna(),["isbn13","description"]].reset_index(drop=True)
```

```python
for i in tqdm(range(0,len(missing_categories))):
    sequence = missing_categories["description"][i]
    predicted_categories.append([generate_predictions(sequence,fiction_categories)])
    isbns.append([missing_categories["isbn13"][i]])
```

```python
missing_predicted_df = pd.DataFrame({"isbn13":isbns,"predicted_categories":predicted_categories})
```

```python
books["isbn13"] = books["isbn13"].astype(str)
missing_predicted_df["isbn13"] = missing_predicted_df["isbn13"].astype(str)
```

```python
books = books.merge(
    missing_predicted_df,
    on="isbn13",
    how="left"
)
books["simple_categories"]=np.where(books["simple_categories"].isna(),books["predicted_categories"],books["description"])
books = books.drop(columns=["predicted_categories"])

category_mapping = {'Fiction': "Fiction",
                    'Juvenile Fiction': "Children's Fiction",
                    'Biography & Autobiography': "Nonfiction",
                    'History': "Nonfiction",
                    'Literary Criticism': "Nonfiction",
                    'Philosophy': "Nonfiction",
                    'Religion': "Nonfiction",
                    'Comics & Graphic Novels': "Fiction",
                    'Drama': 'Fiction',
                    'Juvenile Nonfiction': "Children's Nonfiction",
                    'Science': "Nonfiction",
                    'Poetry': "Fiction"}

books["simple_categories"] = books["categories"].map(category_mapping)

```
