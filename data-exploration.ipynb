```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(),cbar = False,ax = ax)
plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.show()
```

```python
import numpy as np
books["missing_description"]= np.where(books["description"].isna(),1,0)
books["age_of_book"]= 2025- books["published_year"]
```

```python
columns_of_interest = ["num_pages","age_of_book","missing_description","average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman")
#spearman - binary values, pearson - continuous values

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix,annot= True,fmt= ".2f",cmap="coolwarm",cbar_kws={"label":"Spearman Correlation"})
plt.show()
```

```python
books_no_nulls = books[~books['description'].isna()&
                      ~books['num_pages'].isna()&
                      ~books['average_rating'].isna()&
                      ~books['published_year'].isna()
]
books_no_nulls
```

```python
books_no_nulls["categories"].value_counts().reset_index().sort_values("count",ascending = False)
#Distribution is uneven
```

```python
books_no_nulls.loc[:, "words_in_description"] = (books_no_nulls["description"].astype(str).str.split().str.len()
)
```

```python
books_no_nulls
```

```python
books_no_nulls.loc[books_no_nulls["words_in_description"].between(1,4),"description"]
```

```python
books_no_nulls.loc[books_no_nulls["words_in_description"].between(5,14),"description"]
```

```python
books_no_nulls_25_words = books_no_nulls[books_no_nulls["words_in_description"]>=25]
```

```python
books_no_nulls_25_words
```

```python
books_no_nulls_25_words.loc[:,"title_and_subtitle"]=(
    np.where(books_no_nulls_25_words["subtitle"].isna(),books_no_nulls_25_words["title"],
             books_no_nulls_25_words[["title","subtitle"]].astype(str).agg(": ".join,axis=1))
)
```

```python
books_no_nulls_25_words
```

```python
books_no_nulls_25_words.loc[:,"tagged_description"]= (
    books_no_nulls_25_words[["isbn13","description"]].astype(str).agg(": ".join,axis=1))
```

```python
(books_no_nulls_25_words.drop(["subtitle","missing_description","age_of_book","words_in_description"],axis=1)
            .to_csv("books_cleaned.csv",index=False)
 )
```
